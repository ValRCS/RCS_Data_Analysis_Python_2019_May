```python
for i, n in enumerate(range(1,101), start=1):
    print("Index ", i, end=" ")
    if n % 5 == 0 and n % 7 == 0:
        print("FizzBuzz")
    elif n % 5 == 0:
        print("Fizz")
    elif n % 7 == 0:
        print("Buzz")
    else:
        print(n)
```

```python
def fizzBuzz(a=5,b=7, beg=1, end=100):
    '''
    industrial FizzBuzz function
    a is 5 by default
    '''
    for i, n in enumerate(range(beg,end+1), start=beg):
        print("Index ", i, end=" ")
        if n % a == 0 and n % b == 0:
            print("FizzBuzz")
        elif n % a == 0:
            print("Fizz")
        elif n % b == 0:
            print("Buzz")
        else:
            print(n)
```

```python
fizzBuzz(2,3,end=20)
```

```python
fizzBuzz()
```
